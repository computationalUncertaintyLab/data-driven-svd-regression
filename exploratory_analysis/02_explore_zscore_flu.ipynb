```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../analysis_data/hosps_pop_zscore.csv")
df["date"] = pd.to_datetime(df["date"])

states = {
    "06": "California",
    "02": "Alaska"
}

plt.figure(figsize=(10, 5))

for code, name in states.items():
    sub = df[df["location"] == code]
    plt.plot(sub["date"], sub["zflu"], label=name)

plt.axhline(0, color="black", linewidth=1, alpha=0.5)
plt.xlabel("Date")
plt.ylabel("Z-score flu hospitalizations")
plt.title("Z-score flu hospitalizations: California vs Alaska")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

After z-score normalization, California and Alaska are centered at zero and have comparable variance. This confirms that z-scoring removes baseline hospitalization differences and allows us to compare relative flu intensity across states.
